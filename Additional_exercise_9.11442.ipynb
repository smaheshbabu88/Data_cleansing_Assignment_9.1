```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
label = pd.DataFrame(list(map(lambda x : iris.target_names[x], iris.target)), columns=['Species'])
iris = pd.concat([data, label], axis=1)
print(iris.head())
```

```python
import seaborn
import matplotlib.pyplot as plt
import numpy as np
```

### 1) Use the distplot() to see the distribution of the SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm features. Plot them as subplots in a single image.

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 10))
seaborn.distplot(iris['sepal length (cm)'], ax=axes[0,0])
seaborn.distplot(iris['sepal width (cm)'], ax=axes[0,1])
seaborn.distplot(iris['petal length (cm)'], ax=axes[1,0])
seaborn.distplot(iris['petal width (cm)'], ax=axes[1,1])
```

### 2) Do a boxplot of all features except 'Species'.

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 10))
seaborn.boxplot(iris['sepal length (cm)'], ax=axes[0,0], orient='v')
seaborn.boxplot(iris['sepal width (cm)'], ax=axes[0,1])
seaborn.boxplot(iris['petal length (cm)'], ax=axes[1,0])
seaborn.boxplot(iris['petal width (cm)'], ax=axes[1,1], orient='v')
```
